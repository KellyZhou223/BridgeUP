```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
#open file
csvFile = open("PracticalSalinityUnits_Dec_2011.csv")
csvReader = csv.reader(csvFile, delimiter= ',')
data_lists = list(csvReader)
```

```python
#number of rows and columns
print(len(data_lists))
print(len(data_lists[0]))

width = len(data_lists[0])
height = len(data_lists)
```

```python
#numpy matrix
grid = np.empty([height, width, 3], dtype=np.uint8) 
```

```python
#for loops and conditionals
all_values = []

for r, row in enumerate(data_lists):
    for c, value in enumerate(row):
        if float(value) < 99999:
               all_values.append(float(value)) 
        if float(value) <= 30: 
               grid[r, c] = [2, 59, 247]
        elif float(value) <= 31: 
               grid[r, c] = [1, 242, 247]
        elif float(value) <= 32: 
               grid[r, c] = [121, 16, 150]
        elif float(value) <= 33: 
               grid[r, c] = [1, 58, 16]
        elif float(value) <= 34: 
               grid[r, c] = [49, 224, 96]
        elif float(value) <= 35: 
               grid[r, c] = [238, 242, 4]
        elif float(value) <= 36: 
               grid[r, c] = [239, 138, 7]
        elif float(value) <= 37: 
               grid[r, c] = [244, 73, 73]
        elif float(value) <= 38: 
               grid[r, c] = [173, 0, 0]
        else:
               grid[r, c] = [0, 0, 0]
        
print(max(all_values))
print(min(all_values))
```

```python
#color map
plt.title("Salinity of Oceans in December 2011")
plt.imshow(grid)
plt.axis('off')
plt.savefig("Salinity of Oceans in December 2011")
plt.show()
```

```python
#import libraries
import pandas as pd
```

```python
#data frame
orig_df = pd.read_csv("Salinity_info_38N_68W.csv")
orig_df.head()
df = orig_df.dropna()
```

```python
#variables
depth = df["Depth "]
temp = df["Temperature (Celsius)"]
sal = df["Salinity (PSU)"]
den = df["Density (kg/m3)"]

print(max(depth))
print(min(depth))
print(max(temp))
print(min(temp))
print(max(sal))
print(min(sal))
print(max(den))
print(min(den))
```

```python
#bar graph
plt.clf()
plt.plot(depth, temp, color = "teal")
plt.twinx()
plt.plot(depth, den, color = "pink")
plt.title("Temperature and Density of the Ocean in Various Depths")
plt.xticks(range(0, 4001, 500))
plt.xlim(0, 4000)
plt.savefig("Temp and Den")
plt.show()
```

```python
plt.clf()
plt.plot(depth, temp, color = "teal")
plt.legend()
plt. plot(depth, sal, color = "purple")
plt.legend()
plt.title("Temperature and Salinity in the Ocean at Various Depths")
plt.ylim(0, 40)
plt.yticks(range(0, 40, 4))
plt.xlim(0, 4001)
plt.xlabel("Depth")
plt.savefig("Temp and Sal")
plt.show()
```
